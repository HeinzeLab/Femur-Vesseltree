# Determine the number of branches from each branching point

Input: Skeleton-csv files exported from Fiji skeleton analysis

Output: csv-file containing a list of branching degree (0 -> 7)

Note: Only branches longer than 25 um are considered!

```python
# Import all required libraries
import pandas as pd
import numpy as np
import collections
from collections import Counter
import os
```

```python
# Defining the files path 
file_directory= os.listdir()
file_directory= [k for k in file_directory if k.find('.csv') != -1]
```

```python
for k in file_directory:
    shaft_df= pd.read_csv(k)
    
    # Filter data where 'Branch length' >= 25
    filtered_shaft_df= shaft_df[shaft_df['Branch length'] >= 25]
    startingpoints_bin= []
    endpoints_bin= []
    for i in filtered_shaft_df.iloc():
        # Extract V1 and V2 coordinates- V1 coordinates are stored in a bin titled 'startingpoints'
        # and V2 coordinates are stored in a bin titled endpoints.
        startingpoints_bin.append((i['V1 x'], i['V1 y'], i['V1 z']))
        endpoints_bin.append((i['V2 x'], i['V2 y'], i['V2 z']))
    
    # Concatenate V1 and V2 coordinates together in one master list. 
    list_all_points = []
    list_all_points.extend(startingpoints_bin)
    list_all_points.extend(endpoints_bin)
    len(list_all_points)
    
    # Computes a distribution of branching degrees from the master list.
    element_counts= Counter(list_all_points)
    degree_distributions= []
    for i, element in enumerate(element_counts):
        degree_distributions.append(element_counts[element])
    
    # Creates a histogram from the degree distribution.
    result = np.bincount(degree_distributions, minlength = 8)
    
    # Saves the histogram in a csv format. 
    results_dic= {'0': [result[0]], '1': [result[1]], '2': [result[2]], '3': [result[3]], '4': [result[4]], '5': [result[5]], '6': [result[6]], '7': [result[7]]}
    file_name= k.replace('.csv','_branching_points.csv')
    save_dir= os.path.abspath('branching_points')
    file_name= save_dir+'/'+file_name
    results_dataframe= pd.DataFrame(results_dic)
    results_dataframe.to_csv(file_name)
    
    
    
```
