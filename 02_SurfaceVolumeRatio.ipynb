# Extract sum of MK volume and surface

Input: csv-files of a list of surface & volume of all objects in the CD105 or MK segmentation exported from Fiji/3D ROI manager
Output: TXT-file listing the filename, volume sum, surface sum and surface/volume ratio

```python
# 1. Import all required libraries
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

```python
# 2. folder to be processed
path = '*.csv'
save_file_as = 'MK_Surface_Volume.txt'
```

```python
# Initialize empty list of parameter to be exported
list_filenames = list()
list_sum_surface = list()
list_sum_volume = list()
list_ratio = list()
```

```python
# 4. loop over the folder to process all files
for file in glob.glob(path):
    # Get data
    filename = os.path.abspath(file).split(".")[0]
    data = np.genfromtxt(file ,skip_header=1, delimiter=',', usecols=(4,6))
    # Column 4: Vol (unit)
    # Column 6: Surface (unit)
    # Attention: Make sure that these column are in identical order (Settings in 3DROI manager measurements)
    
    # Get sum of volume and surface area
    sum_volume = np.sum(data[:, 0])
    sum_surface = np.sum(data[:, 1])
    ratio = sum_surface / sum_volume
    
    # Append all parameter to a growing list
    list_filenames.append(str(filename))
    list_sum_volume.append(sum_volume)
    list_sum_surface.append(sum_surface)
    list_ratio.append(ratio)
    
header = 'Filename\tVolume[µm³]\tSurface[µm²]\tRatio Surface/Volume'
    
# save results as txt file
np.savetxt(
    save_file_as,
    np.vstack(
        [
            list_filenames,
            list_sum_volume,
            list_sum_surface,
            list_ratio
        ],
    ).T,
    delimiter='\t', fmt="%s", header=header
)                 
```
